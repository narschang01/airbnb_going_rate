```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor, AdaBoostRegressor
from sklearn.svm import LinearSVC, SVC, SVR

from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn import metrics

from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('./data/bos_cleaned_allcolumns.csv')
df.set_index('id', inplace=True)

pd.options.display.max_columns = 100
pd.options.display.max_rows = 3000
pd.options.display.max_colwidth = 200
```

```python
df.head(1)
```

```python
df = df[df['price'] <= 2000]
```

```python
pre = df[df['dataset'] == 'pre-covid']
post = df[df['dataset'] == 'post-covid']
```

```python
pre.head(1)
```

```python
pre.shape
```

```python
post.head(1)
```

```python
post.shape
```

### Define feature

```python
features = list(df.columns.drop(['name', 'host_is_superhost', 'host_identity_verified', 'nb', 'property',
       'room', 'amenities', 'dataset', 'price']))
```

```python
features
```

```python
X_pre = pre[features]
y_pre = pre['price']
```

```python
X_pre_train, X_pre_test, y_pre_train, y_pre_test = train_test_split(X_pre, y_pre, random_state = 42)
```

```python
X_post = post[features]
y_post = post['price']
```

```python
X_post_train, X_post_test, y_post_train, y_post_test = train_test_split(X_post, y_post, random_state = 42)
```

### Scale the data

```python
sc = StandardScaler()
```

```python
sc.fit(X_pre_train)
```

```python
z_pre_train_sc = sc.transform(X_pre_train)
z_pre_test_sc = sc.transform(X_pre_test)
```

```python
z_pre_train_sc.shape, y_pre_train.shape
```

```python
sc.fit(X_post_train)
```

```python
z_post_train_sc = sc.transform(X_post_train)
z_post_test_sc = sc.transform(X_post_test)
```

```python
z_post_train_sc.shape, y_post_train.shape
```

### PCA: Pre-covid

```python
pca = PCA()
```

```python
pca.fit(z_pre_train_sc)
```

```python
Z_pre_train = pca.transform(z_pre_train_sc)
```

```python
Z_pre_test = pca.transform(z_pre_test_sc)
```

```python
var_exp = pca.explained_variance_ratio_
print(f'Explained variance (first 20 components): {var_exp[:50]}')

print('')

# Generate the cumulative explained variance.
cum_var_exp = np.cumsum(var_exp)
print(f'Cumulative explained variance (first 20 components): {cum_var_exp[0:50]}')
```

```python
# Instantiate PCA with 1500 components.
pca = PCA(n_components = 51)

# Fit PCA to training data.
pca.fit(z_pre_train_sc)
```

```python
# Transform Z_train and Z_test.
Z_pre_train = pca.transform(z_pre_train_sc)
Z_pre_test = pca.transform(z_pre_test_sc)
```

```python
# Instantiate linear regression model.
lm = LinearRegression()

# Fit on Z_train.
lm.fit(Z_pre_train, y_pre_train)

# Score on training and testing sets.
print(f'Training Score: {round(lm.score(Z_pre_train, y_pre_train),4)}')
print(f'Testing Score: {round(lm.score(Z_pre_test, y_pre_test),4)}')
```

```python
# Instantiate linear regression model.
knr = KNeighborsRegressor()

# Fit on Z_train.
knr.fit(Z_pre_train, y_pre_train)


# Score on training and testing sets.
print(f'Training Score: {round(knr.score(Z_pre_train, y_pre_train),4)}')
print(f'Testing Score: {round(knr.score(Z_pre_test, y_pre_test),4)}')
```

```python
# Instantiate linear regression model.
knr = KNeighborsRegressor(n_neighbors=8)

# Fit on Z_train.
knr.fit(Z_pre_train, y_pre_train)


# Score on training and testing sets.
print(f'Training Score: {round(knr.score(Z_pre_train, y_pre_train),4)}')
print(f'Testing Score: {round(knr.score(Z_pre_test, y_pre_test),4)}')
```
